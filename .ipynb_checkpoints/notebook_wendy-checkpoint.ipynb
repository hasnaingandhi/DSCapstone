```python
# number installments
installment_score =  orders_payments.groupby('order_id').agg(num_installment=('payment_installments', 'max'), review_score=('review_score', 'mean')).reset_index()
installment_score[['num_installment', 'review_score']].corr().iloc[1, 0]
```

```python
from langdetect import detect
comments = reviews[reviews['review_comment_message'].isna()==False][['review_comment_message', 'review_score']]
comments
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import gensim
from gensim.corpora import Dictionary
from gensim.models import LdaModel
#import pyLDAvis.gensim_models
```

```python
def preprocess_text(text):
    stop_words = set(stopwords.words('portuguese'))
    stop_words.update(('comprei', 'recebi', 'produtos', 'produto', 'compra'))
    stemmer = SnowballStemmer('portuguese')
    tokens = word_tokenize(text.lower())
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return ' '.join(filtered_tokens)

comments['processed'] = comments['review_comment_message'].apply(preprocess_text)
comments_one = comments[comments['review_score']==1]['processed']
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(comments_one)

corpus = gensim.matutils.Sparse2Corpus(X, documents_columns=False)
id2word = Dictionary.from_corpus(corpus, id2word=dict(enumerate(vectorizer.get_feature_names_out())))
```

```python
lda_model = LdaModel(corpus, num_topics=5, id2word=id2word, passes=15)
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```
